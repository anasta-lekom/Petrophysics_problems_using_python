```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OrdinalEncoder
```

```python
t_102 = pd.read_csv('t102.csv', sep=';')
t_103 = pd.read_csv('t103.csv', sep=';')

# в изначальном файле глубины записаны снизу вверх, инверсируем глубины для удобства
t_102 = t_102[::-1]
t_103 = t_103[::-1]
```

Используем в качестве буквы для алфавита описание породы из столбца Rock. Для этого с помощью OrdinalEncoder присвоим каждой "букве" свой шифр в виде целого числа

Проверим, что уникальные "буквы" для обоих скважин совпадают

```python
t_103.Rock.unique() == t_102.Rock.unique()
categories = t_102.Rock.unique()[::-1]
```

"Буквы" совпадают, переходим к кодировке

```python
encoder = OrdinalEncoder(categories=[categories])

t_102['Rock_encoded'] = encoder.fit_transform(t_102[['Rock']])
t_103['Rock_encoded'] = encoder.transform(t_103[['Rock']])
```

Изобразим разрезы

```python
depths = [t_102['Depth'].to_numpy(), t_103['Depth'].to_numpy()]
lith = [(t_102['Rock_encoded'] + 1).to_numpy(), (t_103['Rock_encoded'] + 1).to_numpy()]

fig, axes = plt.subplots(1, 2, figsize=(10, 15))

lith_colors = {
    1: 'black',   # Угли
    2: 'deepskyblue',     # Известняк
    3: 'limegreen',     # Глина
    4: 'yellow'     # Песчаник
}

lith_names = {
    'Угли': 'black',  
    'Известняк': 'deepskyblue',     
    'Глина': 'limegreen',    
    'Песчаник': 'yellow'     
}


for ax, depth, lith_data, well_name in zip(axes, depths, lith, ['Скважина 102', 'Скважина 103']):

    ax.step(lith_data, depth, where='post', color='None', linewidth=0.5)
    
    for i in range(len(depth)-1):
        ax.fill_betweenx(
            y=[depth[i], depth[i+1]],  # Вертикальный диапазон
            x1=0,                     # Левая граница (начало оси X)
            x2=lith_data[i],          # Правая граница (значение литологии)
            color=lith_colors.get(lith_data[i], 'white'),
            step='post',              # Совмещение со ступенькой
            alpha=0.7                 # Прозрачность
        )


    ax.set_title(well_name)
    ax.set_xlabel('Код литологии')
    ax.set_ylabel('Глубина, м')
    ax.set_xticks(lith_data)
    ax.set_ylim(max(depth), min(depth)) 
    ax.grid(True, linestyle='--', alpha=0.5)
    all_depths = depths[0].tolist() + depths[1].tolist()
    ax.set_ylim(max(all_depths), min(all_depths))
   

legend_elements = (
    [Patch(facecolor=color, label=name) for name, color in lith_names.items()]
)
ax.legend(handles=legend_elements, title='Литология и скважины', loc='lower right')

plt.tight_layout()
plt.show()
```

Построим матрицу расстояний Левенштайна

Зададим фунцию, которая будет рассчитывать элементы матрицы расстояний по алгоритму Вагнера — Фишера

```python
def levenstein(matrix, i, j):
    """функция получает i, j индекс элемента в нулевой матрице с количеством строк, равному длине слова А, 
    и количесвтом столбцов, равному длине слова Б;
    
    возвращает значение i, j-элемента матрицы, рассчитанного по алгоритму Вагнера — Фишера"""
    if i == 0 and j == 0:
        return 0
    elif j == 0:
        return i
    elif i == 0:
        return j
    else:
        a = matrix[i][j - 1] + 1  
        b = matrix[i - 1][j] + 1  
        c = matrix[i - 1][j - 1] + int(lith[0][i - 1] != lith[1][j - 1])  
        return min(a, b, c)
```

Построим саму матрицу

```python
rows = len(lith[0]) + 1
cols = len(lith[1]) + 1
dist_matrix = np.zeros((rows, cols), dtype=int)

for i in range(rows):
    for j in range(cols):
        dist_matrix[i][j] = levenstein(dist_matrix, i, j)

print('Матрица расстояний')
print(dist_matrix)
print(f"\nРасстояние Левенштейна: {dist_matrix[rows-1][cols-1]}")
```

Изобразим редакционное предписание с наикратчайшим путем изменения слова

```python
def prescription(matrix, i, j):
    """ вход: i, j-элемент матрицы расстояний \\
        вывод: значение i, j-элемента, \\
        состояние пласта в скважине по сравнению с соседней скважиной,\\
        координты точки (для отображения наикратчайшего пути на редакционном предписании)

    """
    actions = [
               'пласты коррелируют',
               "пласт есть в 102 скважине, выклинивается в 103", 
               "пласт есть в 103 скважине, выклинивается в 102"
               
               ]
    
    values = [
        (matrix[i - 1][j - 1], (i - 1, j - 1)),   # корреляция
        (matrix[i - 1][j], (i - 1, j)),          # выклинивание в 103
        (matrix[i][j - 1], (i, j - 1)),          # выклинивание в 102
        
        ]

    full_info = [(val, action, coords) for (val, coords), action in zip(values, actions)]
    return min(full_info, key=lambda x: x[0]) 
```

Посмотрим на взаимоотношение пластов в разных скважинах и получим координаты для отображения на редакционном предписании

```python
rows = 13
cols = 16
acts = []
coords = [(13, 16)]
while rows > 0 or cols > 0:
    act = prescription(dist_matrix, rows, cols)[1]
    coord = prescription(dist_matrix, rows, cols)[2]
    acts.append(act)
    coords.append(coord)
    rows, cols = coord
    
coords = coords[::-1]
print('Соотношение пластов в 102 и 103 скважинах:')
acts
```

```python
y = [point[0] for point in coords]
x = [point[1] for point in coords]
fig, ax = plt.subplots(figsize=(10, 8))

# Основной график - heatmap
plt.imshow(dist_matrix)
plt.scatter(x, y, color='r')

plt.yticks(np.arange(0, 14))
plt.xticks(np.arange(0, 17))
for i in range(14): 
    for j in range(17): 
        plt.annotate(str(dist_matrix[i][j]), xy=(j, i),
                     ha='center', va='center', color='white') 

plt.suptitle('Редакционное предписание')
plt.tight_layout()
plt.show()
```

Приступим к межскважинной корреляции 

```python
# из результатов выше видно, что некоторых пластов нет в 102 скважине
# создадим маску, чтобы выделить пласты 102 скважины, которые коррелируют с пластами 103 скв
acts = np.array(acts[::-1])
mask = (acts == 'пласты коррелируют')
mask
```
